```python
import time
import pandas as pd
import haversine as hs
from csv import reader
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
website = "https://www.asianpaints.com/store-locator.html"
path= Service('C:/Users/khale/Downloads/chromedriver_win32')
driver = webdriver.Chrome(service=path)
driver.get(website)
```

To store the data of the store name, address, phone number, timings and coordinates, lists are used. And to store all the lists, a data frame is used.

```python
name= []
address= []
timings= []
coordinates= []
phone= []
df = pd.DataFrame({'name': name, 'address': address, 'timings': timings, 'coordinates': coordinates, 'phone number': phone})
```

Two loops are used. One to iterate over all the pin codes and the other nested loop to iterate over all the stores in a particular pin codes. After iterating through all stores in a pin code, the data is appended to the data frame df.

```python
with open('district_pincodes.csv', 'r') as read_obj:
    pincodes = reader(read_obj)
    i=0
    j=0
    
    for pincode in pincodes:
        
        print("For pincode: {} --- S.No: {}".format(pincode,j))
        j+=1
        
        text_box = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "search-input")))
        text_box.clear()
        text_box.send_keys(pincode)
        text_box.submit()
        time.sleep(5)
        
        divlist = driver.find_elements(By.XPATH, "//div[@class='primary-layout_listview_lists newVarientForMap']")
        
        for div in divlist:
            
            try:
                str_name = div.find_element(By.XPATH, ".//h4[@class='d-md-block storeBlock']").text
            except:
                name.append(None)
                str_name = None
            else:
                name.append(str_name)
            
            print("{}--  {}".format(i,str_name))
            i+=1
            
            try:
                latlong_element = div.find_element(By.XPATH, ".//span[@class='spriteIcon-AprevampPhase3 directionIco']")
            except:
                coordinates.append(None)
            else:
                latlong = latlong_element.get_attribute('data-directionurl')
                latlong = latlong[(latlong.rindex("/")+1):]
                coordinates.append(latlong)
            
            try:
                add= div.find_element(By.XPATH, ".//p[@class='description']").text
            except:
                address.append(None)
            else:
                address.append(add)
            
            try:
                ph = div.find_element(By.XPATH, ".//span[@class='open-close d-none d-md-inline-block ml-4']")
            except:
                phone.append(None)
            else:
                phone.append(ph.text[4:])
            
            try:
                store_time = div.find_element(By.XPATH, ".//span[@class='open-close js-open-close']")
            except:
                timings.append(None)
            else:   
                timings.append(store_time.text)
                
        
        df1 = pd.DataFrame({'name': name, 'address': address, 'timings': timings, 'coordinates': coordinates, 'phone number': phone})
        df = df.append(df1, ignore_index = True)
        name[:]= []
        address[:]= []
        timings[:]= []
        coordinates[:]= []
        phone[:]= []

            

driver.quit()
```

```python
df= df.drop_duplicates()
```

Duplicates are deleted and the table is stored in a CSV file. File name has been changed so that the original file remains unchanged.

```python
df.to_csv('asianpaints_data_scraping_copy.csv', index= False)
```

To calcualte 10 stores nearest to the user, user's coordinates are required and then the distance between the stores and user is calcualed and is saved in a CSV file. File name has been changed so that the original file remains.
Sample input for coordinates- Latitude:28.644800, Longitude:77.216721

```python
df3 = pd.DataFrame()
df3 = df

print("Sample input- Latitude:28.644800, Longitude:77.216721")
lat= float(input("Enter latitude"))
long= float(input("Enter longitude"))

loc1=(lat,long)

def dist(loc):
    loc2 = tuple(map(float, loc.split(',')))
    return hs.haversine(loc1,loc2)

dist_list = map(dist, df['coordinates'])
df3['distance'] = list(dist_list)

df3 = df3.round({'distance': 4})
df3.sort_values(by=['distance']).head(10)
df3.sort_values(by=['distance']).head(10).to_csv('nearest_stores_copy.csv', index= False)
```
